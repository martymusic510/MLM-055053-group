```python
# Confidence Interval for the Mean
def confidence_interval(data, confidence=0.95):
    mean = data.mean()
    std = data.std()
    n = len(data)
    z = stats.norm.ppf(1 - (1 - confidence) / 2)  # z-score for 95% CI
    margin_of_error = z * (std / np.sqrt(n))
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error
    return (lower_bound, upper_bound)

ci_results = {}
for col in ar53_non_cat_vars:
    ci_results[col] = confidence_interval(ar53_sample[col])


print("\nConfidence Intervals (95%):")
for col, ci in ci_results.items():
    print(f"{col}: {ci[0]:.2f} to {ci[1]:.2f}")
```

```python
# Test of Homogeneity {Chi-sq} on categorical data

ar53_contingency_table = pd.crosstab(ar53_sample['Import_Export'], ar53_sample['Category'])

# Perform the Chi-square test
chi2, p, dof, expected = chi2_contingency(ar53_contingency_table)

print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of freedom: {dof}")
print("Expected frequencies:")
print(expected)

# Interpret the results
ar53_alpha = 0.05  # Significance level

if p < ar53_alpha:
    print("Reject the null hypothesis: There is a significant association between the two categorical variables.")
else:
    print("There is no significant association between the two categorical variables.")
```

```python
# Test of Normality {Shapiro-Wilk | Kolmogorov-Smirnov | Anderson-Darling | Jarque-Bera

def normality_tests(data):

    ar53_results = {}

    # Shapiro-Wilk test
    shapiro_statistic, shapiro_pvalue = shapiro(data)
    ar53_results["Shapiro-Wilk"] = {"statistic": shapiro_statistic, "p-value": shapiro_pvalue}

    # Kolmogorov-Smirnov test (against a normal distribution)
    ks_statistic, ks_pvalue = kstest(data, 'norm')
    ar53_results["Kolmogorov-Smirnov"] = {"statistic": ks_statistic, "p-value": ks_pvalue}

    # Anderson-Darling test
    anderson_result = anderson(data)
    ar53_results["Anderson-Darling"] = {
        "statistic": anderson_result.statistic,
        "critical_values": anderson_result.critical_values,
        "significance_level": anderson_result.significance_level
    }

    # Jarque-Bera test
    jb_statistic, jb_pvalue = jarque_bera(data)
    ar53_results["Jarque-Bera"] = {"statistic": jb_statistic, "p-value": jb_pvalue}

    return ar53_results
```

```python
from scipy import stats

# Test of Correlation (T-test)
def t_test_correlation(data1, data2):
    t_statistic, p_value = stats.ttest_ind(data1, data2)
    return {"t-statistic": t_statistic, "p-value": p_value}

for col in ar53_non_cat_vars:
    print(f"Normality tests for {col}:")
    normality_test_results = normality_tests(ar53_sample[col])
    for test_name, result in normality_test_results.items():
        print(f"  {test_name}: {result}")

    # Example of T-test
    if len(ar53_non_cat_vars) > 1:
        for i in range(len(ar53_non_cat_vars)):
            for j in range(i+1, len(ar53_non_cat_vars)):
                col1, col2 = ar53_non_cat_vars[i], ar53_non_cat_vars[j]
                t_test_result = t_test_correlation(ar53_sample[col1], ar53_sample[col2])
                print(f"T-test Correlation between {col1} and {col2}: {t_test_result}")
    print()
    print('*'*20)
    print()
```

```python
import time

def time_model_run(model, data):
    start_time = time.time()
    model.fit(data)  # or model.fit_predict(data) depending on the model
    end_time = time.time()
    return end_time - start_time


# Example usage for KMeans (adapt for other models):
from sklearn.cluster import KMeans
kmeans_model = KMeans(n_clusters=2, init='random', random_state=5337) #Example
kmeans_time = time_model_run(kmeans_model, ar53_ds_subset)
print(f"KMeans runtime: {kmeans_time:.4f} seconds")

# Example usage for DBSCAN
from sklearn.cluster import DBSCAN
dbscan_model = DBSCAN(eps=0.5, min_samples=5) #Example
dbscan_time = time_model_run(dbscan_model, ar53_processed)
print(f"DBSCAN runtime: {dbscan_time:.4f} seconds")

#Example usage for Birch
from sklearn.cluster import Birch
birch_model = Birch(n_clusters=2) #Example
birch_time = time_model_run(birch_model, ar53_processed)
print(f"Birch runtime: {birch_time:.4f} seconds")
```

### **Insight:**

This suggests Birch's tree-based structure incurs greater computational cost for the given dataset compared to the centroid-based K-Means or density-based DBSCAN. When **time efficiency is a priority,** **K-Means and DBSCAN** appear to be more favorable choices for this particular analysis.

### **Memory Utilised**

```python
import tracemalloc

def get_model_memory_usage(model, data):
    tracemalloc.start()
    model.fit(data)  # or model.fit_predict(data) depending on the model
    current, peak = tracemalloc.get_traced_memory()
    tracemalloc.stop()
    return peak / (1024 * 1024)  # Convert bytes to MB


# Example usage for KMeans
kmeans_model = KMeans(n_clusters=2, init='random', random_state=5337)
kmeans_memory = get_model_memory_usage(kmeans_model, ar53_ds_subset)
print(f"KMeans memory usage: {kmeans_memory:.2f} MB")

# Example usage for DBSCAN
dbscan_model = DBSCAN(eps=0.5, min_samples=5)
dbscan_memory = get_model_memory_usage(dbscan_model, ar53_processed)
print(f"DBSCAN memory usage: {dbscan_memory:.2f} MB")

# Example usage for Birch
birch_model = Birch(n_clusters=2)
birch_memory = get_model_memory_usage(birch_model, ar53_processed)
print(f"Birch memory usage: {birch_memory:.2f} MB")
```

### **> Observation:**
Birch demonstrated significantly higher memory usage (196.31 MB) compared to K-Means (0.32 MB) and DBSCAN (2.33 MB).

### **> Insight:**
This substantial difference indicates Birch's tree-based structure requires considerable memory to store and manage its internal data representations. K-Means and DBSCAN, with their lower memory footprints, offer better scalability for larger datasets or memory-constrained environments. This reinforces Birch's potential limitation when dealing with extensive data volumes.
